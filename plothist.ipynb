```python
#open 0.7_0.7777777_n1001_nologit.csv using csv
import csv
with open('0.7_0.7777777_n1001_nologit.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
data=[float(n[0]) for n in data if n[0]!="" and n[0]!="\n" and n[0]!="None"]

```

```python
#plot data as a hist from 0.7 to 0.7777777 with 10 bins
import numpy as np
import matplotlib.pyplot as plt

# Create normalized histogram
plt.hist(data, bins=10, edgecolor='black', alpha=0.7, density=True)

# Title and labels
plt.title('Histogram of Data, floats from 0.7 to 0.7777777, n= '+str(len(data)))
plt.xlabel('Value')
plt.ylabel('Density')

# Show plot
plt.show()
```
