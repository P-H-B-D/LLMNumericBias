```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filename = "ratings_comparison_no_skew_110.csv"
df = pd.read_csv(filename)
df = df.dropna()

# Calculate the histogram data for both sets of ratings without plotting
true_hist = np.histogram(df['true_rating'], bins=range(1, 7))
openai_hist = np.histogram(df['openai_rating'], bins=range(1, 7))

# Determine the maximum bin count between the two histograms
max_count = max(max(true_hist[0]), max(openai_hist[0]))

# Set up the figure and axes
fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

# Plot the true ratings histogram
ax[0].hist(df['true_rating'], bins=range(1, 7), edgecolor="k", align='left', rwidth=0.8)
ax[0].set_title("True Ratings")
ax[0].set_xlabel('Value')
ax[0].set_ylim([0, max_count + 5])  # Add a little extra room at the top

# Plot the OpenAI ratings histogram
ax[1].hist(df['openai_rating'], bins=range(1, 7), edgecolor="k", align='left', rwidth=0.8)
ax[1].set_title("OpenAI Ratings")
ax[1].set_xlabel('Value')

plt.tight_layout()
plt.show()
```
